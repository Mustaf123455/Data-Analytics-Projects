```python

with open("insurance.csv","r+") as insurance:
    insurance = insurance.readlines()
    def all_Categories_list_retriever(file):
     return file[0].strip().split(',')
    all_categories_list = all_Categories_list_retriever(insurance)
    print (all_categories_list)
    def separate_list_retriever(file,category):
        result = []
        for line in file:
         result.append(line.strip().split(",")[all_categories_list.index(category)])
        return result
    sexes = separate_list_retriever(insurance,'sex')
    bmis = separate_list_retriever(insurance,'bmi')
    childrens = separate_list_retriever(insurance,'children')
    smokers = separate_list_retriever(insurance,'smoker')
    regions =separate_list_retriever(insurance,'region')
    chargess= separate_list_retriever(insurance,'charges')
    ages = separate_list_retriever(insurance,'age')
    def list1_list2_relation(list_1,list_2):  
        final_list = []
        for item1,item2 in zip(list_1,list_2):
         final_list.append(item1)
         final_list.append(item2)
        return final_list
    final_list = list1_list2_relation(sexes,chargess)
    print(final_list)
    male = []
    female = []
    for item in range(0,len(final_list)):
     if final_list[item] == 'female':
         female.append(float(final_list[item+1]))
     elif final_list[item] == 'male':
         male.append(float(final_list[item+1]))
    average_male = sum(male)/len(male)
    average_female  = sum(female)/len(female)
    print(average_male)
    print(average_female)
    print("Males tend to spend more on medical insurance with an amount of %s" % average_male)
    final_list = list1_list2_relation(bmis,chargess)
    integer_bmi =[]
    for item in range(1,len(bmis)):
        integer_bmi.append(float(bmis[item]))
    average_bmi = sum(integer_bmi)/len(integer_bmi)
    print(average_bmi)
    print(final_list)
    higher_bmi =[]
    lower_bmi =[]
    for item in range(2,len(final_list),2):
      if float(final_list[item])>average_bmi:
          higher_bmi.append(float(final_list[item+1]))
      elif float(final_list[item])<average_bmi:
          lower_bmi.append(float(final_list[item+1]))
    average_higher_bmi = sum(higher_bmi)/len(higher_bmi)
    average_lower_bmi = sum(lower_bmi)/len(lower_bmi)
    print(average_higher_bmi)
    print(average_lower_bmi)
    print("people with higher bmis then the average payed more on insurance costs with an amount of %s"%average_higher_bmi)
    final_list = list1_list2_relation(ages,chargess)
    print(final_list)
    integer_age =[]
    for item in range(2,len(final_list),2):
        integer_age.append(float(final_list[item]))
    average_age = sum(integer_age)/len(integer_age)
    lower_average = []
    higher_average = []
    for item in range(2,len(final_list),2):
        if float(final_list[item])>average_age:
         higher_average.append(float(final_list[item+1]))
        elif float(final_list[item])<average_age:
         lower_average.append(float(final_list[item+1]))
    average_lower_average = sum(lower_average)/len(lower_average)
    average_higher_average = sum(higher_average)/len(higher_average)
    print (average_lower_average)
    print (average_higher_average)
    print("people with a higher age had more insurance costs with an amount of %s" % average_higher_average)
    final_list = list1_list2_relation(childrens,chargess)
    zero_children =[]
    one_child = []
    two_children = []
    three_children = []
    four_children= []
    more_than_four_children = []
    for item in range(2,len(final_list),2):
     if float(final_list[item]) == 0:
         zero_children.append(float(final_list[item+1]))
     elif float(final_list[item]) == 1:
         one_child.append(float(final_list[item+1]))    
     elif float(final_list[item]) == 2:
         two_children.append(float(final_list[item+1]))
     elif float(final_list[item]) == 3:
         three_children.append(float(final_list[item+1]))
     elif float(final_list[item]) == 4:
         four_children.append(float(final_list[item+1]))
     elif float(final_list[item])>4:
         more_than_four_children.append(float(final_list[item+1]))  
    average_zero_children = sum(zero_children)/len(zero_children)
    average_one_child = sum(one_child)/len(one_child)
    average_two_children = sum(two_children)/len(two_children)
    average_three_children = sum(three_children)/len(three_children)
    average_four_children = sum(four_children)/len(four_children)
    average_more_than_four_children = sum(more_than_four_children)/len(more_than_four_children)
    print (average_zero_children)
    print (average_one_child)
    print (average_two_children)
    print (average_three_children)
    print (average_four_children)
    print (average_more_than_four_children)
    print("people with three children tend to payer higher medical insurance costs with an amount of %s" %average_three_children)
    final_list = list1_list2_relation(smokers,chargess)
    non_smoker_list =[]
    smoker_list =[]
    for item in range(2,len(final_list),2):
        if final_list[item] == 'yes':
            smoker_list.append(float(final_list[item+1]))
        elif final_list[item] == 'no':
            non_smoker_list.append(float(final_list[item+1]))
    average_non_smoker =sum(non_smoker_list)/len(non_smoker_list)
    average_smoker =sum(smoker_list)/len(smoker_list)
    print(average_non_smoker)
    print(average_smoker)
    print("The average smoker payed more with an amount of %s"%average_smoker)
    final_list = list1_list2_relation(regions,chargess)
    northeast= []
    northwest =[]
    southeast=[]
    southwest=[]
    for item in range(2,len(final_list),2):
        if final_list[item] == 'southeast':
            southeast.append(float(final_list[item+1]))
        elif final_list[item] == 'northeast':
            northeast.append(float(final_list[item+1]))
        elif final_list[item] == 'northwest':
            northwest.append(float(final_list[item+1]))
        elif final_list[item] == 'southwest':
            southwest.append(float(final_list[item+1]))
    average_north_west = sum(northwest)/len(northwest)
    average_north_east = sum(northeast)/len(northeast) 
    average_south_west = sum(southwest)/len(southwest)
    average_south_east = sum(southeast)/len(southeast)     
    print(average_south_west)
    print(average_north_west)
    print(average_south_east)
    print(average_north_east)
    print("The people in the south east payed the most amount of insurance with an amount of %s"%average_south_east)

            
    
    
    
    
        
          
        
        
    
     
    

  
   




         
         
```
